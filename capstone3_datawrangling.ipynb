```python
# Although this data is readily available, I wanted to create the bands to practice.
def bollinger_bands(stock_df):
    import numpy as np
    import pandas as pd
    import matplotlib as plt
    
    df = stock_df[['AMZN Adj Close']]
  
    
    # Create rolling average and standard deviation
    sma = df.rolling(window=20).mean().dropna()
    
    rstd = df.rolling(window=20).std().dropna()
    # Create upper band and lower bands
    upper_band = sma + 2*rstd
    lower_band = sma - 2*rstd

    #Create columns with upper and lower band data
    upper_band = upper_band.rename(columns = {'AMZN Adj Close': 'Upper BB'})
    lower_band = lower_band.rename(columns = {'AMZN Adj Close': 'Lower BB'})
    bb = df.join(sma, rsuffix = 'SMA')
    bb = bb.join(upper_band).join(lower_band)
    bb = bb.dropna()
         
    bb.columns = ['AMZN Adj Close', 'SMA', 'BB Upper Band', 'BB Lower Band']
    band_conditions = [
        (bb['SMA'] >= bb['BB Upper Band']),
        (bb['SMA'] < bb['BB Upper Band']) & (bb['SMA']> bb['BB Lower Band']),
        (bb['SMA'] <= bb['BB Lower Band'])
    ]
    band_categories = ['Overbought', 'In Range', 'Oversold']
    bb['Bollinger_Status'] = np.select(band_conditions, band_categories)
    bb = bb.drop(['AMZN Adj Close'], axis = 1)
    return bb
    
 
bollinger_data = bollinger_bands(df_amzn)
```

```python
def MACD(stock_df):
    import numpy as np
    k = stock_df['AMZN Adj Close'].ewm(span = 12, adjust = False, min_periods=12).mean()
    d = stock_df['AMZN Adj Close'].ewm(span=26, adjust = False, min_periods = 26).mean()
    macd = k-d
    macd_trigger =  macd.ewm(span = 9, adjust = False, min_periods = 9).mean()
    macd_c_d = macd - macd_trigger
    df_macd = pd.concat([macd, macd_trigger, macd_c_d], axis = 1)
    df_macd.columns =  ['MACD', 'MACD_Trigger', 'MACD_Con/Div']
    
    macd_conditions = [
        (df_macd['MACD_Con/Div'] > 0),
        (df_macd['MACD_Con/Div'] < 0)
    ]
    macd_categories = ['Bullish', 'Bearish']
    df_macd['MACD_Market'] = np.select(macd_conditions, macd_categories)
    df_macd = df_macd.dropna()
    return df_macd
    
    
df_macd = MACD(df_amzn)
```

```python
# Sentiment analysis for Amazon stock
def Sentiment_stock(stock):
    import os
    import matplotlib.pyplot as plt
    from bs4 import BeautifulSoup
    import pandas as pd
    import nltk
    from urllib.request import urlopen, Request
    from nltk.sentiment.vader import SentimentIntensityAnalyzer
    
    news_tables = {}
    url = 'https://finviz.com/quote.ashx?t='
    full_url = url + stock
    req = Request(url = full_url, headers = {"User-Agent": "Chrome"})
    response = urlopen(req)
    html = BeautifulSoup(response, 'html.parser')
    news_table = html.find(id = 'news-table')
    news_tables[stock] = news_table
    
    amzn = news_tables[stock]
    amzn_tr = amzn.findAll('tr')
    
    for x, table_row in enumerate(amzn_tr):
        a_text = table_row.a.text
        td_text = table_row.td.text
        if x == 3:
            break
    news_list = []
    for file_name, news_table in news_tables.items():
        for i in news_table.find_all('tr'):
            text = i.get_text()
            date_scrape = i.td.text.split()
            if len(date_scrape) == 1:
                time = date_scrape[0]
            else:
                date = date_scrape[0]
                time = date_scrape[1]
            tick = file_name.split('_')[0]
            
            news_list.append([tick, date, time, text])
            
    vader = SentimentIntensityAnalyzer()
    columns = ['ticker', 'date', 'time', 'headline']
    news_df = pd.DataFrame(news_list, columns = columns)
    scores = news_df['headline'].apply(vader.polarity_scores).tolist()
    scores_df = pd.DataFrame(scores)
    news_df = news_df.join(scores_df, rsuffix = '_right')
    news_df['date'] = pd.to_datetime(news_df.date).dt.date
    day_score_df = news_df.groupby(['date']).agg({'compound': 'mean'}).reset_index()
    return day_score_df
    

day_score_df = Sentiment_stock('AMZN')
```

```python
def RSI(stock_df, lower, upper):
    # This function will take amazon stock and return a Relative String Index data to be added to the main dataframe.
    import numpy as np
    df = stock_df[['AMZN Adj Close']]
    
    change = df['AMZN Adj Close'].diff()
    low = lower
    high = upper
    up = change.copy()
    down = change.copy()
    up[up<0] = 0
    down[down>0] = 0
    
    avg_up = up.rolling(14).mean()
    avg_dwn = down.rolling(14).mean().abs()
    
    rsi = 100 * avg_up / (avg_up + avg_dwn)
    df_s = [df, rsi]
    df = pd.concat(df_s, axis = 1)
    df.reset_index()
    df.columns= ['AMZN Adj Close', 'RSI']
    rsi_conditions = [
        (df['RSI'] > high),
        (df['RSI']<= high) & (df['RSI'] >= low),
        (df['RSI'] < low)
    ]
    rsi_categories = ['Bullish', 'Normal', 'Bearish']
    df['RSI_Status'] = np.select(rsi_conditions, rsi_categories)
    df = df.dropna()
    df = df.drop(['AMZN Adj Close'], axis = 1)
    final = [df, low, high]
    
    return final
df_RSI = RSI(df_amzn, 30,70)
```
